```python
import numpy as np
from matplotlib import pyplot as plt
x = np.arange(1,11)
```

```python

y = x*2
y2 = x*3
y2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)

plt.title("line plot")
plt.xlabel("x_label")
plt.ylabel("y_label")
plt.show
```

```python
plt.plot(x,y, color='green', linestyle=':', linewidth=2)

plt.title("line plot")
plt.xlabel("x_label")
plt.ylabel("y_label")
plt.show
```

```python
plt.plot(x,y)
plt.plot(x,y2, color='green', linestyle=':', linewidth=2)

plt.title("line plot")
plt.xlabel("x_label")
plt.ylabel("y_label")
plt.show
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, color='blue', linestyle=':', linewidth=3)

plt.subplot(1,2,2)
plt.plot(x,y2, color='green', linestyle='-', linewidth=2)
plt.show
```
